# Kodun içinde kullanım

```python
import matplotlib.pyplot as plt
```

Basit bir kullanım için NumPy arrayları ile paslaşabiliriz. İki np.arrayı verip bunu çiz dediğimizde matplotlib bunu hızlıca yapar

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

# Şekilli Şukullu Grafikler yapmak

İstersek çizgi istersek  nokta yapıyorduk. Peki ya ikisini de aynı anda istersek?
Bu sefer dümdüz "o"  yazmak yerine marker="o" yazdık.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker = "o")
plt.show()
```

```python
plt.plot(ypoints, marker = "*")
plt.show()
```

Bakalım aşağıdaki kod ne yapacak?

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

noktalarla işaretli kırmızı kesikli çizgili bir grafik oldu.
bu 'o:r' ifadesinden geliyor.           
Sıralama : marker|line|color şeklinde           

Marker tipleri yukarıda verilmişti.

### Çizgi tipleri

':'	Dotted line

marker boyutu ile oynamak için "marker size" olarak ms yazacağız

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

daha da ileri gidelim ve "marker edge color" ekleyelim. Bunu mec şeklinde ifade edeceğiz.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

Marker rengi içinde marker face color manasında mfc yazacağız.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

çizgi şeklini line style olarakta ifade edebiliriz.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

renk  değişiklieri için ise color parametresi var

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

Kalınlık için ise linewidth seçeneği mevcut.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

# Eksenlere İsim Vermek

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

İsim vermenin yanı sıra farklı fontlarda kullanabilirsin

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1,loc = 'left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

Bir color map ekleyip farklı color değerleri vermek mümkündür. Bu sanki 3 eksen varmış gibi çalışmanı sağlar.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

Yada farklı sizeler oluşturabilirsin.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

İkisini combolarsak çılgın bir şey olur

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Histogramlar

Histogramlar çok sayıda ölçümün dağılımını göstermek için iyidir. Aşağıda  x = np.random.normal(170, 10, 250) ile 250 rastgele sayı üreteceğiz. Ortalamaları 170 ve standart sapma 10 olacak. Her çalıştırdığında rastgele bir dağılım olacak. Olay çok sayıda verinin dağılımını görmek.

```python
import matplotlib.pyplot as plt
import numpy as np

# ortlaması 170 olan standart sapması 10 olan 250 sayı üretir
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```
